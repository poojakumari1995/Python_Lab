```python
# Note: If the number of elements in the queue is less than k, 
# then print “There are not enough elements in the queue”.




def print_element(q,k):
    # q- input queue, k- kth element from the front
    # print the kth element from the front end of queue
    # write your code here. 
    if q.size()<k:
        print ("There are not enough elements in the queue")
    else:
        elm=q.size()-k
        for i in range(elm+1):
            a=q.dequeue()
        print(a)
```

```python
# Sort in Increasing and Decreasing Order

l= ast.literal_eval(input())
odd=[]
even=[]
for i in range(len(l)):
    if i%2==0:
        even.append(l[i])
    else:
        odd.append(l[i])
even.sort()
odd.sort(reverse=True)
even.extend(odd)
print(even)
```

```python
# function to detect duplicate_paranthesis       
def duplicate_paranthesis(string):
  #write your logic here
	Stack = []
	for ch in string:
		if ch == ')':
			top = Stack.pop()
			elementsInside = 0
			while top != '(':
			
				elementsInside += 1
				top = Stack.pop()
			
			if elementsInside < 1:
				return True
		else:
			Stack.append(ch)
	
	return False
  
  
```
